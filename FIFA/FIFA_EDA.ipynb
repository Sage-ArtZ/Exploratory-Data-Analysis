```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
fifa = pd.read_csv('players_20.csv')
```

```python
fifa.head()
```

```python
for col in fifa.columns:
    print(col)
```

```python
fifa['nationality'].value_counts()
```

```python
fifa['nationality'].value_counts().head(5)
```

```python
plt.figure(figsize=(15,5))
plt.bar(fifa['nationality'].value_counts().head(5).keys(),fifa['nationality'].value_counts().head(5),color='red',width=0.5)
```

```python
player_salary = fifa[['short_name','wage_eur']]
```

```python
player_salary = player_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
player_salary.head(10)
```

```python
plt.figure(figsize=(15,5))
plt.bar(player_salary['short_name'].head(5),player_salary['wage_eur'].head(5),color='blue',width=0.5)
```

```python
player_value = fifa[['short_name','value_eur']].sort_values(by=['value_eur'],ascending=False)
```

```python
player_value.head(10)
```

```python
plt.figure(figsize=(15,5))
plt.bar(player_value['short_name'].head(5),player_value['value_eur'].head(5),color=['red','orange','pink','cyan','blue'],width=0.5)
```
